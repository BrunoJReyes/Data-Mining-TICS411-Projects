#K-means

```python
data = data[['energy', 'tempo', 'speechiness', 'acousticness', 'instrumentalness', 'danceability', 'loudness', 'valence']]
data.sample(5)
```

- Eleccion de hiperparametros de K-means

 - Para evaluar el numero optimo de clusters es necesario computar las distancas de los datos en el grafico en general que se ve y despues minimizarlo eligiendo la cantidad de clusters, para hacer accionar el valor de k es necesario aplicar una heuristica, es decir, aumentar el numero de clusters hasta que la diferencia en la funcion de error (la suma de todos los within clusters distance se estabilice), donde podemos notar que el optimo es 7.

```python
import numpy as np
import plotnine as ptn
from plotnine import *
```

```python
from sklearn.cluster import KMeans
```

```python
sse = [] #Variable para guardar el within cluster distance
numK = 20
for k in range(1, numK): #nÃºmero de cluster a probar
    #Creando el modelo, entrenandolo y obteniendo el valor del within cluster distance
    kmeans = KMeans(n_clusters=k,n_init=10)
    kmeans.fit(data)
    sse.append(kmeans.inertia_)
#graficando

print(sse)
tempDataFrame=pd.DataFrame(range(1, numK),columns=["numK"])
tempDataFrame["sse"]=sse
```

```python
(ggplot(tempDataFrame)+aes(x="numK",y="sse")+theme_bw()+geom_line()+labs(x="NÃºmero de clusters",y="WCD")
  +scale_x_continuous(breaks=range(1,numK))
  )
```

Coeficiente de Sillhuete

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() #Creamos un objeto de la clase StandardScaler
scaled_features = scaler.fit_transform(data) #est
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score
maxK=10
sse = np.zeros(maxK-2) # NO USAR []
CS = np.zeros(maxK-2) # NO USAR []
for k in range(3, maxK):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(scaled_features)
    sse[k-2]=kmeans.inertia_ #NO USAR sse.append(kmeans.inertia_)
    silhouette_avg = silhouette_score(scaled_features, kmeans.labels_)
    CS[k-2]=silhouette_avg #NO USAR CS.append(silhouette_avg)
    print("calculamos:",k)
    
#Generating the dataframa
tempData=pd.DataFrame(range(3, maxK),columns=["K"])
tempData["sse"]=sse
tempData["CS"]=CS
print(ggplot(tempData)+aes(x="K",y="sse")+theme_bw()+geom_line()+labs(x="NÃºmero de clusters",y="WCD")+scale_x_continuous(breaks=range(3,maxK)))
print(ggplot(tempData)+aes(x="K",y="CS")+theme_bw()+geom_line()+labs(x="NÃºmero de clusters",y="Coeficiente de Silhouette promedio")+scale_x_continuous(breaks=range(3,maxK)))    
```

#DBSCAN

```python
import pandas as pd
import numpy as np

data = pd.read_csv("beats.csv")
```

```python
data = data[['energy', 'tempo', 'speechiness', 'acousticness', 'instrumentalness', 'danceability', 'loudness', 'valence']]
data.sample(5)
```

```python
import numpy as np
import plotnine as ptn
from plotnine import *
```

- Eleccion de hiperparametros DBSCAN

  - MinPts: Se escogio dependiando de la cantidad de las columnas de nuestras variables, a nuestro numero determinado de columnas le sumamos 1.

  - Eps: Para calcular su eps optimo, se procede a usar un metodo de nos calcula la distancia promedio entre cada punto y sus k vecinos mas cercanos, donde k son los MinPts que ya definimos, el cual se definio como 0.6.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() 
norm_data = scaler.fit_transform(data)
```

```python
#### ElecciÃ³n de eps ####
from sklearn.neighbors import NearestNeighbors
from matplotlib import pyplot as plt

neighbors = NearestNeighbors(n_neighbors = 9)
neighbors_fit = neighbors.fit(norm_data)
distances, indices = neighbors_fit.kneighbors(norm_data)

distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.plot(distances)
```

```python
sep = np.arange(len(distances))


(ggplot(data, aes(x = sep, y = distances)) +
  scale_x_continuous(breaks = range(0,700000,13000))+
 scale_y_continuous(breaks = np.arange(0,2.5,0.20))+
 theme(axis_text_x=element_text(rotation=90, hjust=1))+
  geom_line())
```

```python
#Escalando los datos
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() #Creamos un objeto de la clase StandardScaler
scaled_features = scaler.fit_transform(data) #Transformamos los datos
```

```python
from sklearn.cluster import DBSCAN

df_cluster = data

dbscan = DBSCAN(eps=0.6, min_samples=9, algorithm='brute', p=0.9)

preds = dbscan.fit_predict(df_cluster)
```

```python
#nuesttro sihouette score (coeficiente de silueta) es malo 

silhouette_score(df_cluster, preds)
```

- La metrica de Sihouette nos da un valor de -0.7231092852822228, por ende el modelo no es idoneo para clusterizar.
